```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}
```

```python
def input_invalid(command, commands):
    """Функция проверки правильности введенной команды"""""
    
    if command not in commands:
        print("Введена несуществующая команда. Проверьте ввод")
        return True
    else:
        return False
```

```python
def get_doc_info():
    """Функция получения информации о документе по его номеру"""
    
    doc_numb = input("Введите номер документа:")
    doc_info = [i for i in documents if i["number"] == doc_numb]
    if doc_info!=[]:
        return doc_info[0]
    else:
        print("Документ не найден в базе")
        print_documents()
        return 
```

```python
def get_shelf_info(necessity):
    """Функция получения информации о полке по ее номеру. Параметр necessity отвечает, 
    должен ли полка существовать во время поиска, или, наоборот, мы ждем, что такой полки в словаре не будет """
    
    shelf_numb = input("Введите номер полки:")
    if shelf_numb in directories.keys():
        if necessity == True:
            return shelf_numb
        else:
            print("Такая полка уже существует. Текущий список полок:", directories.keys())
            return 
    else:
        if necessity == True:
            print("Такой полки не существует. Текущий список полок:", directories.keys())
            return 
        else:
            return shelf_numb
```

```python
def search_shelf_for_doc(doc_numb):
    """Функция поиска полки по номеру документа"""
    
    shelf_numb = [i for i in directories.keys() if doc_numb in directories[i]]
    return shelf_numb[0]
```

```python
def check_shelf_emptiness(shelf_numb):
    """Функция проверки, пустая ли полка, или на ней есть документы """
    
    if directories[shelf_numb] != []:
        print("На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ", directories.keys())
        return False
    else:
        return True
```

```python
def print_documents():
    """Функция вывода списка документов """
    
    print("Текущий список документов:")
    for i in documents:
        print("№: {}, тип: {}, владелец: {}, полка хранения: {}".format(i["number"],i["type"],i["name"], search_shelf_for_doc(i["number"])))
```

```python
available_commands = ["q","p","s","d","i","as","ds","ad"]
command = input("Введите команду: ")

while(command != "q"):
    if input_invalid(command,available_commands):
        command = input("Введите команду: ")
        continue
        
    else:
        if command == "p" or command =="s" or command =="d":
            doc = get_doc_info()
            if(doc!=None):
                if command == "p":
                    print("Владелец документа:", doc["name"])
                elif command == "s":
                    shelf = search_shelf_for_doc(doc["number"])
                    print("Документ хранится на полке:", shelf)
                else:
                    shelf = search_shelf_for_doc(doc["number"])
                    directories[shelf].remove(doc["number"])
                    documents.remove(doc)
                    print("Документ удален")
                    print_documents()    
                    
        elif command == "i":
            print_documents()
            
        elif command == "as":
            shelf = get_shelf_info(False)
            if shelf!=None:
                directories[shelf] = []
                print("Полка добавлена. Текущий перечень полок: ", directories.keys())
                
        elif command == "ds":
            shelf = get_shelf_info(True)
            if shelf != None:
                if check_shelf_emptiness(shelf)==True:
                        del directories[shelf]
                        print("Полка удалена. Текущий перечень полок: ", directories.keys())
                        
        elif command == "ad":
            doc_numb = input("Введите номер документа:")
            doc_type = input("Введите тип документа:")
            doc_name = input("Введите владельца документа:")
            shelf = get_shelf_info(True)
            if shelf!=None:
                documents.append({'number':doc_numb, 'type': doc_type , 'name':doc_name})
                directories[shelf].append(doc_numb)
                print("Документ добавлен.")
                print_documents()
            
        command = input("Введите команду: ")
        
```
