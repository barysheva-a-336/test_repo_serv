## Задача 2:
Дана переменная, в которой хранится четырехзначное число (год). Необходимо написать программу, которая выведет, является ли данный год високосным или обычным.

Пример работы программы:

year = 2020
Результат:
Високосный год

year = 2019
Результат:
Обычный год

### Решение:

```python
year = int(input()) 
if year%4 == 0 and (
    year%100!=0 or (year%100 == 0 and year%400 == 0)
):
    print("Високосный")
else:
    print("Обычный")
```

## Задание 3
Необходимо написать программу, которая будет запрашивать у пользователя месяц и дату рождения и выводить соответствующий знак зодиака.

Пример работы программы:

Введите день:
30

Введите месяц:
Август

Результат:
Ваш знак зодиака: Дева

Введите день:
29

Введите месяц:
Октябрь

Результат:
Ваш знак зодиака: Скорпион

### Решение:

```python
'''В решении участвуют список и словарь. Список нужен для сохранения названий месяцев,словарь - для сохранения данных о 
датах действия определенного знака гороскопа.
В словаре есть лишняя запись "Козерог_1" - для правильного определения Козерогов, т.к. у них происходит переход с 12го месяца 
на 1й.
Чтобы в результате выводить правильное название знака зодиака, после поиска по словарю - лишнюю запись обрабатываю на выходе.
+
Программа не производит проверку значений, поданных на вход.'''
months = ["январь", "февраль","март","апрель","май","июнь","июль","август","сентябрь","октябрь","ноябрь","декабрь"]
zodiac = {
    "Овен":["03-21","04-20"],
    "Телец":["04-21","05-20"],
    "Близнецы":["05-21","06-21"],
    "Рак":["06-22","07-22"],
    "Лев":["07-23","08-23"],
    "Дева":["08-24","09-23"],
    "Весы":["09-24","10-23"],
    "Скорпион":["10-24","11-22"],
    "Стрелец":["11-23","12-21"],
    "Козерог":["12-22","12-31"],
    "Козерог_1":["01-01","01-20"],
    "Водолей":["01-21","02-20"],
    "Рыбы":["02-21","03-20"],
}

day = input("Введите день:")
if (len(day) == 1):
    day = "0" + day
month = str(months.index(input("Введите месяц:").lower())+1)
if(len(month) == 1):
    month = "0" + month
date = month + '-' + day
zodiac_sign = [k for k in zodiac.keys() if (zodiac[k][0] <= date <= zodiac[k][1])]
if "Козерог_1" in zodiac_sign:
    zodiac_sign[0] = "Козерог"
print("Ваш знак зодиака:",*zodiac_sign)
```

## Задание 5 (необязательное)
Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу, которая будет определять, является ли данный билет “счастливым”. Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.

Примеры работы программы:

number = 123456
Результат:
Неасчастливый билет

number = 123321
Результат:
Счастливый билет

### Решение:

```python
ticket_number = list(input())
ticket_number = [int(i) for i in ticket_number]
if(sum(ticket_number[0:3]) == sum(ticket_number[3:6])):
    print("Счастливый билет")
else:
    print("Несчастливый билет")
```

## Задание 6 (необязательное)
Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник). Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

если пользователь выбрал круг, запрашиваем его радиус,
если треугольник – длины трех его сторон;
если прямоугольник – длины двух его сторон.
Пример работы программы:

Введите тип фигуры:
Круг

Введите радиус круга:
10

Результат:
Площадь круга: 314.16

Введите тип фигуры:
Треугольник

Введите длину стороны A:
2

Введите длину стороны B:
2

Введите длину стороны C:
3

Результат:
Площадь треугольника: 1.98

### Решение:

```python
figures = {
    'треугольник': lambda x, y, z: (((x+z+y)/2)*(((x+z+y)/2)-x)*(((x+z+y)/2)-y)*(((x+z+y)/2)-z))**(1/2),
    'прямоугольник': lambda x, y, z: x*y,
    'круг': lambda x, y, z: 3.14159*(x**2),
}
figure = input("Введите тип фигуры: ").lower()
y, z = 0, 0
if figure == 'круг':
    x = int(input("Введите радиус круга: "))
else:
    x = int(input("Введите длину стороны A: "))
    y = int(input("Введите длину стороны B: "))
    if figure == 'треугольник':
        z = int(input("Введите длину стороны C: "))
print("Площадь " + figure + 'а: ', round(figures[figure](x,y,z),2))
```
